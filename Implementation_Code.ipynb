```python
import numpy as np
import pandas as pd
import yfinance as yf
import scipy as sp
import statsmodels.tsa.stattools as ts
import matplotlib.pyplot as plt
import datetime
```

```python
def  plot_pairs(data1,data2):
    fig,(ax1,ax2)=plt.subplots(2)
    fig.suptitle("pair of stocks")
    ax1.plot(data1)
    ax2.plot(data2)
    plt.show()
def cointigrate(data1,data2):
    result=sp.stats.linregress(data1,data2)
    residuals=data1-data2*result.slope#I(0) SERIES
    adf=ts.adfuller(residuals)
    return adf[0]
def dnld(stocks ,s_date,e_date,df):
    df=yf.download(stocks,start=s_date,end=e_date)['Adj Close']
    return df
```

```python
No_trading_days = 252
start_date = datetime.datetime(2017, 12, 31)
end_date = datetime.datetime(2020, 12, 31)
stocks = ["INDUSINDBK.NS", "HDFCBANK.NS", "PNB.NS", "ICICIBANK.NS", "SBIN.NS", "IDFCFIRSTB.NS", "KOTAKBANK.NS", "FEDERALBNK.NS", "AXISBANK.NS", "AUBANK.NS"]
df=pd.DataFrame()
stock_value = dnld(stocks, start_date, end_date,df)
conf99 = []
conf95 = []
conf90 = []
for i in stock_value:
    for j in stock_value:
        if(i!=j):
            result = ts.coint(stock_value[i],stock_value[j])
            if result[1] <0.01:
                conf99.append([i,j])
            elif result[1]<0.05:
                conf95.append([i,j])
            elif result[1]<0.1:
                conf90.append([i,j])
print("pairs with 99% confidence lvl",conf99)
print("pairs with 95% confidence lvl",conf95)
print("pairs with 90% confidence lvl",conf90)
```

```python
def residual(data1,data2):
    result=sp.stats.linregress(data1,data2)
    rf=data2-data1*result.slope
    return rf
def slope(data1,data2):
    result=sp.stats.linregress(data2,data1)
    return result.slope
def plotxy(data1,data2):
    x=np.zeros(len(data1))
    for i in range(len(data1)):
        x[i]=i+1
    y=residual(data1,data2)
    plt.plot(x,y)
    plt.show()
def std_div(data1,data2):
    residuals=residual(data1,data2)
    std=np.std(residuals)
    return std
def zscore(series):
    lookback = 20
    zscore = []
    for i in range (20,len(series)):
        period = series[i - 20:i]
        zscore.append((series[i] - period.mean())/ period.std())
    return zscore
```

```python
Data1=conf99[0][0]
Data2=conf99[0][1]
CAP=100000
b_start="2020-07-01"
b_end="2023-07-01"
dfd=pd.DataFrame()
dfd=dnld(conf99[0],b_start,b_end,dfd)
residu=residual(dfd[Data1],dfd[Data2])
res=np.array(zscore(residu))
price_x=dfd[Data1][20:]
price_y=dfd[Data2][20:]
thresold_0=1.0
thresold_1=-1.0
beta=slope(dfd[Data1],dfd[Data2])
print(residu.size)
```

```python
cap=np.zeros(residu.size-20)
cap[0]=100000
position=np.zeros(residu.size-20)
idlewindow=0   #to decrease the trading frequency so that we do not exit in case of any sudden jump 
no_of_trades=0
for  i in range(1,residu.size-20):
    if(res[i-1]>=thresold_0 and position[i-1]==0 and idlewindow>3):
        position[i]=1
        invest_x=(-cap[i-1]/(beta-1))    #sort selling the above one
        invest_y=(beta*cap[i-1])/(beta-1)       #buying the lower one
        vol_x=invest_x/price_x[i-1]
        vol_y=invest_y/price_y[i-1]
        cap[i]=vol_x*price_x[i]+vol_y*price_y[i]
        idlewindow=1
        no_of_trades+=1
    if(res[i-1]<=4*thresold_1 or res[i-1]>=4*thresold_0):#for risk managememt 
        position[i]=0
        cap[i]=cap[i-1]
    elif (res[i-1]<=0 and position[i-1]==1 and idlewindow>3):
        position[i]=0
        cap[i]=cap[i-1]
        idlewindow=0
    elif (res[i-1] <= thresold_1 and position[i-1]==0 and idlewindow>2):
        position[i]=-1 
        invest_x=(-cap [i-1] /(beta-1))  #sort selling the above one
        invest_y=beta*cap[i-1]/(beta-1)      #buying the lower one
        vol_x=invest_x/price_x[i-1]
        vol_y=invest_y/price_y[i-1]
        cap[i]=vol_x*price_x[i]+vol_y*price_y[i]
        idlewindow=1
        no_of_trades+=1
    elif (res[i-1]>=0 and position[i-1]==-1 and idlewindow>6):
        position[i]=0
        cap[i]=cap[i-1]
        idlewindow=0
    elif(position[i-1]==1 or position[i-1]==-1):
        cap[i]=vol_x*price_x[i]+vol_y*price_y[i]
        position[i]=position[i-1]
        idlewindow+=1
    else :
        cap[i]=cap[i-1]
        idlewindow+=1
```

```python
def max_drawdown(cap):
    new_cap=[]
    for i in cap:
        new_cap.append(i-100000)
    max_drawdown=min(new_cap)/1000
    return -1*max_drawdown
```

```python
def sharpe(cap):
    daily_ann_sharpe=[]
    risk_free_return=5.75/252
    new_cap=[]
    for i in cap:
        new_cap.append(i-100000)
    for i in new_cap:
        daily_ann_sharpe.append(((i)-risk_free_return*1000)/np.std(new_cap))
    mean_sharpe=np.mean(daily_ann_sharpe)
    return mean_sharpe
```
